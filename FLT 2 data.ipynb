```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
FLT_2 = pd.read_excel('FLT 2 datasets.xlsx')
FLT_2
```

```python
fig, ax = plt.subplots(7,1, figsize=(5, 10), dpi = 200)
count = 0
for data in FLT_2.columns:
    if data != "t(s)":
        ax[count].plot(FLT_2["t(s)"], FLT_2[data])
        ax[count].set_title(data)
        count += 1
plt.tight_layout()
plt.show()
```
