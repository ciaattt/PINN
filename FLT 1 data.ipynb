```python
FLT_1 = pd.read_excel('FLT 1 datasets.xlsx')
FLT_1
```

```python
fig, ax = plt.subplots(7,1, figsize=(5, 10), dpi = 200)
count = 0
for data in FLT_1.columns:
    if data != "t(s)":
        ax[count].plot(FLT_1["t(s)"], FLT_1[data])
        ax[count].set_title(data)
        count += 1
plt.tight_layout()
plt.show()
```
