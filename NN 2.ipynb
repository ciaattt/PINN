```python
import tensorflow as tf
from tensorflow import keras
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.model_selection
import sklearn.metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import numpy as np
```

```python
x = pd.read_excel ("FLT 1 Revised.xlsx").drop (columns='t(s)')
x = x[0:(len(x)-1)].to_numpy()
x.shape
```

```python
pd.read_excel ("FLT 1 Revised.xlsx").drop (columns='true speed (m/s)').drop (columns='t(s)')
```

```python
y = pd.read_excel ("FLT 1 datasets.xlsx").drop (columns='t(s)').drop (columns='engine inclinataion (deg)')
y = y[1:(len(y))].to_numpy()
y.shape
```

```python
pd.read_excel ("FLT 1 datasets.xlsx").drop (columns='true speed (m/s)').drop (columns='t(s)').drop (columns='engine inclinataion (deg)')
```

```python
model = keras.Sequential([
    keras.layers.Dense(7, input_shape =(7,),),
    keras.layers.Dense(9, activation = "sigmoid"),
    keras.layers.Dense(6, activation = "linear")
])
model.compile(optimizer = keras.optimizers.SGD(learning_rate=0.05, momentum = 0.03),
              loss = 'MSE',
               metrics = ['RootMeanSquaredError']
            )

model.fit(x,y, epochs = 1000, batch_size = 100)
model.summary()
```

```python
y_t = model.predict(x)
```
