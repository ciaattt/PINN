```python
def y (a,c):

    y_in = []
    x_in = [0,1,2,3,4]

    for x in x_in:
        y_in.append (
            a*x + c
        )  
    print (y_in)
    return x_in, y_in
x_in, y_in = y (2,3)
```

```python
def Y_1 (w_1, B_1):

    Y_1 = []
    x_in

    for x in x_in:
        Y_1.append (
            w_1*x + B_1
        )
    return x_in,Y_1
x_in, Y_1 = Y_1 (6,4)
print (x_in)
print (Y_1)
```

```python
def Y_2 (w_2,B_2):

    Y_2 = []
    Y_1 

    for x_in2 in Y_1:
        Y_2.append (
            w_2 * x_in2 + B_2
        )
    return Y_1, Y_2
Y_1, Y_2 = Y_2 (4, 2)
print (Y_1)
print (Y_2)
```

```python
def dLdW_2 (w_2, B_2):

    N = 5
    sumdldw2 = 0
    for N in range (N):
        sumdldw2 += (-2/5)*(y_in[N-1] - w_2 * Y_1[N-1] - B_2)*(Y_1[N-1])
    return sumdldw2

sumdldw2 = dLdW_2(4,2)
print (sumdldw2)
```

```python
def dLdB_2 (w_2, B_2):

    N = 5
    sumdldb2 = 0
    for N in range (N):
        sumdldb2 += (-2/5)*(y_in[N-1] - w_2 * Y_1[N-1] - B_2)
    return sumdldb2

sumdldb2 = dLdB_2(4,2)
print (sumdldb2)
```
